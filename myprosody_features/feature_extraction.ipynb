```python
import myprosody as mysp
import pandas as pd
```

```python
class Paths:
  meta_data = '/vol/tensusers3/jdusseljee/2324-asr/myprosody_features/meta_data.csv'
  concatenated = '/vol/tensusers3/jdusseljee/2324-asr/myprosody_features/full'
```

```python
def get_features(row):
  p=row["ID"]

  return pd.Series(mysp.myspsr(p, Paths.concatenated), index=[f'f{n+1}' for n in range(15)])
```

```python
train_df = pd.read_csv(Paths.meta_data)
```

```python
train_df[[f'f{n+1}' for n in range(15)]] = train_df.apply(get_features, axis=1)
```
